```python
import os
import shutil
import pandas as pd

# Set the paths
image_folder = "images/images"
csv_file = "Train.csv"

# Read the CSV file
df = pd.read_csv(csv_file)

# Create folders for each label if they don't exist
label_folders = set(df['damage'])
for label in label_folders:
    folder_path = os.path.join(image_folder, str(label))
    os.makedirs(folder_path, exist_ok=True)

# Iterate through each row in the CSV file
for index, row in df.iterrows():
    image_filename = row['filename']
    label = row['damage']

    # Construct the source and destination paths
    src_path = os.path.join(image_folder, image_filename)
    dest_path = os.path.join(image_folder, str(label), image_filename)

    # Move the file to the corresponding folder
    shutil.move(src_path, dest_path)

print("Files moved to folders based on labels.")
```
